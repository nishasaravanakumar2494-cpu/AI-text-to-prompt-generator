```python
# aLanguage prompt engiering th multiple templates
from langchain_core.prompts import PromptTemplate
demoPrompt = '''
I need you to act as a language translator for the given input '{sample}' either in hindi tamil like wise the output {text_output}
'''
prompt2 = PromptTemplate(input_variables=["sample",'text_output'],template = demoPrompt)
prompt2.format(sample="how are you",text_output='hindi')
chain2= prompt2 | llm
response = chain2.invoke({
    "sample": "how are you",
    "text_output": "Tamil"
})
print(response.content)
```
